```python
import re

# Check for values in the column that are not integers
non_int_values = [value for value in df['10th Class Science Marks (out of 100)'] if not isinstance(value, int)]

# Filter values that match patterns like '10/10'
special_values = [value for value in non_int_values if re.match(r'^\d+/\d+$', value)]

# Print the result
print(special_values)

```

```python
# Convert special values to percentages and update the dataset
for i, value in df['10th Class Science Marks (out of 100)'].items():
    if isinstance(value, str) and re.match(r'^\d+/\d+$', value):
        numerator, denominator = map(int, value.split('/'))
        df.at[i, '10th Class Science Marks (out of 100)'] = (numerator / denominator) * 100

```

```python
# Define a function to handle special cases and convert to float
def convert_to_float(value):
    if isinstance(value, str):  # Check if the value is a string
        try:
            return float(value)  # Convert regular strings to float
        except ValueError:
            return None  # Handle invalid strings as NaN
    elif isinstance(value, (int, float)):  # Already numeric
        return float(value)
    return None  # Handle other cases like None or non-convertible values

# Apply the function to convert the column
df['10th Class Science Marks (out of 100)'] = df['10th Class Science Marks (out of 100)'].apply(convert_to_float)

# Ensure the column is of float type
df['10th Class Science Marks (out of 100)'] = df['10th Class Science Marks (out of 100)'].astype(float)

```

```python
import pandas as pd

# Load dataset
df = pd.read_csv(r"C:\Users\dell\OneDrive\Desktop\ALL FILE\CGPA-Prediction-Model\Anonymised_2028_Batch_Marks.csv")  # Change this to your actual dataset file

# Specify columns to skip (columns where cleaning should NOT be applied)
skip_columns = ['Gender', '10th Class Board','12th Class School Category','12th Class Board']  # Replace with actual column names

# Define a function to clean values
def clean_value(value):
    if isinstance(value, str):  # Check if the value is a string
        value = value.strip()  # Remove extra spaces
        if '/' in value:  # Convert '9/10' format to float
            num, denom = map(float, value.split('/'))
            return (num / denom) * 100
        elif '%' in value:  # Remove '%' and convert to float
        
            return float(value.replace('%', ''))
        elif ',' in value:
            return float(value.replace(',', '.'))
        elif value.lower() == 'pass':  # Replace 'Pass' with 90
            return 90.0
        elif value.lower() in ['not given', 'na']:  # Handle missing values
            return 0.0
        elif '(ENG)' in value:  # Remove '(ENG)' and convert to float
            return float(value.replace('(ENG)', ''))
        elif 'percentile' in value:  # Remove 'percentile' and convert to float
            return float(value.replace('percentile', ''))
        elif value.replace('.', '', 1).isdigit():  # Convert numeric strings to float
            return float(value)
    return value  # Return original value if it's already a number

# Apply cleaning only to selected columns
df_cleaned = df.copy()  # Create a copy of the original dataset
df_cleaned[df_cleaned.columns.difference(skip_columns)] = df_cleaned[df_cleaned.columns.difference(skip_columns)].applymap(clean_value)

# Save the cleaned dataset
df_cleaned.to_csv("cleaned_data.csv", index=False)  # Saves as CSV

# Display cleaned DataFrame
print(df_cleaned)
```
