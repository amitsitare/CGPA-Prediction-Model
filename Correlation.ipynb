```python
# Load dataset (Replace 'your_dataset.csv' with actual file)
import pandas as pd
df = pd.read_csv(r"C:\Users\dell\OneDrive\Desktop\All file\CGPA-Prediction-Model\cleaned_data.csv")
```

```python
newdf=df.copy()
```

```python
# remove the string value
df_float = newdf.drop(["Gender", "10th Class Board", "12th Class School Category", "12th Class Board"], axis=1, errors='ignore')
```

```python
# correlation matrix of all column
correlation=df_float.corr()
correlation
```

```python
# correlation with CGPA
correlation_cgpa=correlation["CGPA"].sort_values(ascending=False)
correlation_cgpa
```

```python
# bargraph
import matplotlib.pyplot as plt

# Sorting correlation values in descending order
correlation_sorted = correlation["CGPA"].sort_values(ascending=False)

# Plotting the bar chart
plt.figure(figsize=(10, 5))
correlation_sorted.plot(kind='bar', color='skyblue', edgecolor='black')

# Adding labels and title
plt.xlabel("Features")
plt.ylabel("Correlation with CGPA")
plt.title("Correlation of CGPA with Other Columns")
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility

# Show the plot
plt.show()
```
